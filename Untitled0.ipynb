#**PROBABILIDAD Y ESTADISTICA**

**1.2 Apuntes:**
##**¿Qué es la lógica?**
*El proceso de determinar si una afirmación es falsa o correcta a través del rasocinio.*
###¿Para qué me sirve la lógica en la materia?
*Nos sirve para analizar un cierto dato o evento y dar a una mejor organización de estos.*
##**¿Qué es una axioma?**
*Los axiomas son verdades incuestionables universalmente válidas y evidentes.*
##**¿Qué es la probabilidad?**
Es una medida que muestra la posibilidad de que ocurra un evento, nos sirve para indicar la facilidad o la dificultad de que ocurra dicho evento.

**Forma de calcular la probabilidad:**  la probabilidad de que ocurra un evento es igual al número de casos favorables partido por el número total de casos posibles.
##**¿Qué es la estadística?**
Ciencia que estudia, analiza y organiza los datos.

##**"*Concepto de Estadistica Descriptiva"***

La estadistica descriptiva es una rama de la estadistica que se encarga de resumir y escribir las características principales de un conjunto de datos.

1.   **Recolección de datos**
2.   **Calculo de medidadas descriptivas**
3.   **Presentación de datos**
4.   **Identificación de patrones y relaciones**

##**"*Concepto de Estadistica Inferencial"***

En esta existe una emisión de juicios o conclusiones basados en los datos existentes.



*   **Estimación de parámetros**
*   **Prueba de hipótesis**
*   **Intervalos de confianza**
*   **Modelado estadístico**
