#**PROBABILIDAD Y ESTADÍSTICA**

**EJERCICIO DE EJEMPLO:**

1.  Un administrador de red analiza el tráfico generado por direcciones IP en una
red empresarial y detecta que el 30% del tráfico proviene de una única IP.
Conclusión: Se descubrió que esa IP corresponde a un ataque DDoS. Al limitar las conexiones desde esa dirección, se estabilizó el tráfico de la red.

**EXPLICACIÓN DEL COMO SE PUEDE ANALIZAR EL SIGUIENTE CASO EN PROBABILIDAD Y ESTADÍSTICA**


---
*  *PROBABILIDAD*
---

Estamos hablando del tráfico de red, el administrador dispone de varias redes en uno solo, por lo tanto varias "IP"
Ahora hablamos que en una sola IP tenemos un 30%
Lo cuál corresponde a un 0.30 de probabilidad de que el paquete de datos provenga de esa IP
Lo normal sería que las redes se distribuyan de manera equitativa, por lo tanto el tráfico también.

Se llega a la conclusión de que hay una anomalía con base a los datos existentes y los eventos que se esperan, el siguiente analisis pudo revelar que se trataba de un  “ataque distribuido denegación de servicio”.



---
*  *ESTADÍSTICA*
---

En estadística nos basamos en los datos que tenemos de igual manera, si tenemos un 30% en una sola IP, se detecta la anomalía, puesto que debemos analizar que una red de tráfico estable debe haber un equilibrio equitativo entre las redes.

Entonces para poder estabilizar el tráfico y tener un mejor manejo de él se haría un debido recorte o limitación de redes.
